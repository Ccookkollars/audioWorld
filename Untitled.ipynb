```python
import sounddevice as sdd
import struct
import math
import numpy as np
import pandas as pd
import queue
from matplotlib import pyplot as plt
import csv
%matplotlib notebook
```

```python
aud = dict()
aud['samplerate'] = 44100
aud['channels'] = 1
with open(r'/home/caleb/Downloads/002.wav', 'rb') as f:
    #riff
    print(' RIFF, filesize, WAVE')
    print(struct.unpack('4si4s', f.read(12)))
    print('\n subchunk1: fmt, size')
    print(struct.unpack('4si', f.read(8)))
    #format pcm = 1, channels mono = 1
    print(' pcm = 1, channels mono = 1')
    print(struct.unpack('hh', f.read(4)))
    #samplerate
    print(' samplerate, byterate')
    print(struct.unpack('ii', f.read(8)))
    print(' bytes per sample, bits per sample')
    (Bps, bps) = struct.unpack('hh', f.read(4))
    print((Bps, bps))
    
    while True:
        print('\n subchunkid')
        subch_name = struct.unpack('4s', f.read(4))
        subch_name = subch_name[0]
        print(subch_name)
        if subch_name == b'data':
            break;
        #subchunksize
        sz = struct.unpack('i', f.read(4))
        sz = sz[0]
        f.read(sz)
    
    sz = struct.unpack('i', f.read(4))
    adat = np.array(struct.unpack('h'*int(sz[0]/2), f.read(sz[0]))).astype(np.float)/37260
    
```

```python
start = 7
sec = 5
bdat = adat[start*44100:(44100*(start+sec))]
sdd.play(bdat)
print(len(bdat))
print(pd.DataFrame(bdat))
```

```python
notenames = []
notefreqs = []
with open(r'/home/caleb/projects/audioWorld/notes.csv', 'r') as f:
    reader = csv.reader(f)
    for line in reader:
        notenames.append(line[1])
        notefreqs.append(float(line[2]))
print(pd.DataFrame(list(zip(notenames, notefreqs))))
```
